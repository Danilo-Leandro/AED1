# Exercício 1:

```python
numero = int(input("Digite um número"))
if numero % 3 == 0 and numero % 5 == 0:
    print("O número é divisível por 3 e por 5.")
else:
    print("O número não é divisível por ambos.")
```

# Exercício 2:

```python
valor = float(input("Digite o valor da compra"))
if valor <= 1000:
    desconto = valor * 0.10
elif valor <= 5000:
    desconto = valor * 0.20
else:
    desconto = valor * 0.30
valor_final = valor - desconto
print(f"Desconto: R${desconto}")
print(f"Valor final: R${valor_final}")
```

# Exercício 3

```python
largura = float(input("Digite a largura do comodo em metros: "))
comprimento = float(input("Digite o comprimento do como em metros: "))
litros_lata = float(input("Digite a quantidade de litros: "))
altura = 2.8
area_porta = 0.8 * 2.10
rendimento = 3 

area = 2 * altura * (largura + comprimento) - area_porta

litros_necessarios = area / rendimento

latas = litros_necessarios / litros_lata

print(f"Área a ser pintada: {area} m2")
print(f"Litros de tinta necessarios: {litros_necessarios} litros")
print(f"Quantidade exata de latas: {latas}")


```
