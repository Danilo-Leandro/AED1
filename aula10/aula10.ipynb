Perguntas e respostas sobre Tuplas
1. Por que a tupla tem menos métodos que a lista? O que isso revela sobre sua natureza?
# As tuplas possuem menos métodos porque são objetos e imutáveis . Enquanto as listas precisam de métodos para adicionar ( append), remover ( remove) ou ordenar ( sort) elementos, as tuplas não permitem alterações após sua criação. Isso revela que sua natureza é uma estrutura de dados estática, alternativamente como uma "coleção de leitura" estável.

2. Qual é a diferença entre as funções index() e count() em uma tupla?
# O método index()retorna a posição (índice) da primeira ocorrência de um valor específico na tupla, lançando um erro caso o valor não exista. Já o método count()conta e retorna quantas vezes um valor específico aparece naquela tupla, retornando zero se ele não for encontrado.

3. Como posso acessar elementos de uma tupla e o que acontece se eu tentar modificá-los?
# Os elementos são acessados ​​através de colchetes e o índice desejado (ex: tupla[0]). Se você tentar modificar um elemento (ex: tupla[0] = novo_valor), o Python retornará um erro de tipo ( TypeError), informando que o objeto 'tuple' não suporta atribuição de itens.

4. Ao trabalhar com tuplas, percebi que não consegui usar métodos como append() e remove(). Isso é uma limitação ou uma vantagem?
# Pode ser visto como ambos, dependendo do contexto. É uma vantagem quando se busca segurança, pois garante que os dados não serão alterados acidentalmente por outras partes do código. É uma restrição apenas se você precisar de uma estrutura dinâmica que muda de tamanho constantemente, caso a lista seja a escolha correta.

5. Quando devo preferir uma tupla em vez de uma lista? Posso dar exemplos práticos com explicação?
# Você deve preferir tuplas quando os dados são constantes ou representam uma unidade lógica que não deve mudar. Exemplos práticos:
# Coordenadas Geográficas: (longitude, latitude) — um ponto no mapa é fixo;
# Configurações de Sistema: (host, porta, protocolo) — dados que não devem ser alterados durante a execução;
# 3 Retorno de Funções: Quando uma função precisa retornar múltiplos valores de uma vez, ela geralmente os agrupa em uma tupla.

6. Qual é a diferença entre tupla() e declarar uma tupla com parênteses?
# Declarar com parênteses (a, b)é uma sintaxe literal para criar uma tupla diretamente. Já a função tuple()é um construtor que converte outros objetos iteráveis ​​(como uma string, uma lista ou um dicionário) em uma tupla. Por exemplo, tuple([1, 2])transforme uma lista em tupla.

7. Quais são os impactos de usar tuplas em termos de desempenho e segurança em sistemas maiores?
# Desempenho: Tuplas são um pouco mais rápidas que listas e consomem menos memória porque o Python aloca o espaço exato necessário de forma estática.
# Segurança: Em sistemas grandes, o uso de tuplas previne bugs de "efeitos colaterais", onde uma função altera ivertidamente uma lista que é usada em outro lugar do sistema. Além disso, por serem imutáveis, tuplas podem ser usadas como chaves em dicionários, o que não é possível com listas.

# Exercícios


```python
# LINHA 1: Cria uma lista vazia chamada lista_idades
# Por que: essa lista vai guardar todas as idades digitadas pelo usuário
# Exemplo: após digitar 20, 30 e 40 ficará assim -> [20, 30, 40]
lista_idades = []

# LINHA 2: Cria um laço de repetição infinito
# Por que: o programa precisa continuar pedindo idades até o usuário decidir parar
# Ideia: "Repita para sempre, até eu mandar parar"
while True:

    # LINHA 3: Pede uma idade ao usuário, converte para número inteiro e guarda na variável idade
    # Por que: precisamos receber a idade como número, não como texto
    # O input() mostra a mensagem e espera o usuário digitar
    # O int() converte o que foi digitado para número inteiro
    # Exemplo: se o usuário digitar 25, então idade = 25
    idade = int(input("Entre com uma idade: "))

    # LINHA 4: Verifica se a idade digitada é diferente de -1
    # Por que: o valor -1 foi usado como sinal de parada
    # != significa "diferente de"
    # Se a pessoa digitar uma idade normal, o programa guarda
    # Se digitar -1, o programa entende que deve encerrar
    if idade != -1:

        # LINHA 5: Adiciona a idade digitada ao final da lista lista_idades
        # Por que: toda idade válida deve ser armazenada para depois calcular quantidade, soma e média
        # Exemplo: se lista era [20, 30] e idade=40, depois do append fica [20, 30, 40]
        lista_idades.append(idade)

    # LINHA 6: Define o que acontece quando a condição do if for falsa (ou seja, quando idade == -1)
    # Por que: se a idade for -1, ela não deve ser adicionada à lista, e o programa deve parar
    else:

        # LINHA 7: Interrompe o laço while
        # Por que: quando o usuário digita -1, significa que terminou de informar as idades
        break

# LINHA 8: Converte a lista lista_idades em uma tupla e guarda na variável tupla_idades
# Por que: o exercício trabalha com tuplas. A lista foi usada primeiro porque permite
# adicionar elementos com append(). Depois, os dados são transformados em tupla.
# Diferença: lista é mutável (pode mudar), tupla é imutável (não pode ser alterada)
# Exemplo: lista [20, 30, 40] vira tupla (20, 30, 40)
tupla_idades = tuple(lista_idades)

# LINHA 9: Soma todos os valores da tupla
# Por que: precisamos da soma total das idades para calcular a média
# Exemplo: se a tupla for (20, 30, 40), então total = 90
total = sum(tupla_idades)

# LINHA 10: Conta quantos elementos existem na tupla
# Por que: precisamos saber quantas idades foram digitadas para calcular a média
# Exemplo: se a tupla for (20, 30, 40), então qtd = 3
qtd = len(tupla_idades)

# LINHA 11: Calcula a média das idades
# Por que: fórmula da média = soma dos valores / quantidade de valores
# Exemplo: total=90 e qtd=3, então media = 30.0
media = total / qtd

# LINHA 12: Mostra na tela a quantidade de idades digitadas
# Por que: para informar ao usuário quantos valores válidos foram inseridos
# O f antes da string permite colocar variáveis dentro do texto usando {}
# Exemplo: se qtd=3, exibe -> Total de idades digitadas: 3
print(f"Total de idades digitadas: {qtd}")

# LINHA 13: Mostra na tela a média calculada
# Por que: esse é um dos resultados principais do programa
# Exemplo: se media=30.0, exibe -> A média de idades é: 30.0
print(f"A média de idades é: {media}")
```
